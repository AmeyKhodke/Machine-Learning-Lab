```python

print("Simple Calculator")

choice = 0

while True:
    print("\nChoose an operation:")
    print("1. Addition")
    print("2. Subtraction") 
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")
    
    try:
        choice = int(input("Enter your choice: "))
        
        if choice == 5:
            print("Exiting calculator...")
            break
            
        if choice < 1 or choice > 5:
            print("Invalid choice. Please try again.")
            continue
            
        a = int(input("Enter first number: "))
        b = int(input("Enter Second Number: "))
        
        match choice:
            case 1:
                print(f"Result: {a + b}")
            case 2:
                print(f"Result: {a - b}")
            case 3:
                print(f"Result: {a * b}")
            case 4:
                if b == 0:
                    print("Error: Cannot divide by zero!")
                else:
                    print(f"Result: {a / b}")
            
    except ValueError:
        print("Invalid input. Please enter numbers only.")
```

```python
import matplotlib.pyplot as plt

x = list(map(float, input("Enter x coordinates for the graph: ").split()))
y = list(map(float, input("Enter y coordinates for the graph: ").split()))

plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Custom Line Graph')
plt.grid(True)
plt.tight_layout()
plt.show()
```
