```python
data = data.drop(['S.No.'], axis=1)
data.info()
```

```python
from datetime import date
date.today().year
data['Car_Age'] = date.today().year - data['Year']
data.head()
```

```python
# split the data into two parts Brand and Model
data['Brand'] = data.Name.str.split().str.get(0)
data.head()
```

```python
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
data.head()
```

```python
data[['Name','Brand','Model']]
```

```python
# Annual Mileage Calculation
data['Annual_Mileage'] = data['Kilometers_Driven'] / data['Car_Age']
data.head()
```

```python
# Usage intensity of a Car
data['Usage_Intensity'] = pd.cut(data['Annual_Mileage'],
                                 bins = [0,8000,15000,float('inf')],
                                 labels=['Low','Medium','High'])
data.head()
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
category = data['Fuel_Type']
value = data['Fuel_Type'].value_counts().head(10)
plt.bar(value.index, value.values)
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('fuel_type_distribution.png')
plt.show()
```

```python
brand_type = data['Brand']
values = data['Brand'].value_counts().head(10)
plt.bar(values.index, values.values, color='orange')
plt.title('Brand Distribution ')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig('brand_type_distibutin.png', bbox_inches='tight')
plt.show()
```

```python
# Top 5 cars by Price (Pie Chart)
top = data.nlargest(5,'Price')
plt.pie(top['Price'], labels=top['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Cars by Price')
plt.axis('equal')
plt.savefig('top_5_cars_by_price.png',bbox_inches='tight')
plt.show()
```

```python
# Histogram of Car Purchases by year
plt.hist(data['Year'], bins=20, color='magenta', edgecolor='black')
plt.title('Car Purchases by Year')
plt.xlabel('Year')
plt.ylabel('No. of Cars')
plt.savefig('car_purchases_by_year.png', bbox_inches='tight')
plt.show()
```

```python
# Histogram of Car Age
plt.hist(data['Car_Age'], bins=20, color='lime')
plt.title('Car Age Distribution')
plt.xlabel('Car Age')
plt.ylabel('No of Cars')
plt.savefig('car_age_distribution.png', bbox_inches='tight')
plt.show()
```

```python
# Bivariate Analysis of Relationship between Categorical and Continous Variables
fig, axarr = plt.subplots(2,2, figsize=(16,10))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
data.groupby('Fuel_Type')['Price'].mean().plot(kind='bar', ax=axarr[0,0], color='SkyBlue')
axarr[0,0].set_title('Average Price by Fuel Type')
axarr[0,0].set_ylabel('Average Price')
axarr[0,0].tick_params(axis='x', rotation=45)
data.groupby('Brand')['Price'].mean().sort_values(ascending=False).head(7).plot(kind='bar', ax=axarr[0,1], color='orange')
axarr[0,1].set_title('Average Price by Brand')
axarr[0,1].set_ylabel('Average Price')
axarr[0,1].tick_params(axis='x', rotation=45)
data.groupby('Usage_Intensity')['Price'].mean().plot(kind='bar', ax=axarr[1,0], color='lime')
axarr[1,0].set_title('Average Price by Usage Intensity')
axarr[1,0].set_ylabel('Average Price')
axarr[1,0].tick_params(axis='x', rotation=45)
data.groupby('Location')['Price'].mean().plot(kind='bar', ax=axarr[1,1], color='magenta')
axarr[1,1].set_title('Average Price by Location')
axarr[1,1].set_ylabel('Average Price')
axarr[1,1].tick_params(axis='x', rotation=45)
for ax in axarr.flat:
    ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('bivariate_analysis.png', bbox_inches='tight')
plt.show()
```

```python
# Bivariate Analysis using scatter plots
fig, axarr = plt.subplots(2,2, figsize=(14,10))
scatter_style = {'s': 50, 'alpha': 0.6, 'edgecolor': 'w'}
data.plot.scatter(x='Car_Age', y='Price', ax=axarr[0,0], color='SkyBlue',**scatter_style)
axarr[0,0].set_title('Car Age vs Price')
data.plot.scatter(x='Kilometers_Driven', y='Price', ax=axarr[0,1], color='orange',**scatter_style)
axarr[0,1].set_title('Kilometers Driven vs Price')
data.plot.scatter(x='Annual_Mileage', y='Price', ax=axarr[1,0], color='lime',**scatter_style)
axarr[1,0].set_title('Annual Mileage vs Price')
data.plot.scatter(x='Car_Age', y='Kilometers_Driven', ax=axarr[1,1], color='magenta',**scatter_style)
axarr[1,1].set_title('Car Age vs Kilometers Driven')
plt.tight_layout()
plt.savefig('bivariate_scatter_plots.png',bbox_inches='tight')
plt.show()
```

```python
# Single Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(data['Car_Age'], data['Price'], color='skyBlue', s=70, alpha=0.6, edgecolor='black')
plt.title('Car Age vs Price',fontsize=16, fontweight='bold')
plt.xlabel('Car Age (years)', fontsize=13)
plt.ylabel('Price (in Lakhs)', fontsize=13)
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.5)
plt.savefig('car_age_vs_price_scatter.png', bbox_inches='tight')
plt.show()
```
