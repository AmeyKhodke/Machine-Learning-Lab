```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
plt.title("Salary Distribution Plot")
sns.distplot(data['Salary'])
plt.show()
```

```python
plt.scatter(data['YearsExperience'],data['Salary'], color='lightcoral')
plt.title("Experience vs Salary")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
x = data.iloc[:, :1] # Independent
y = data.iloc[:, 1:] # Dependent
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred_test = reg.predict(x_test) # Predicted value of y_test
y_pred_train = reg.predict(x_train) # Predicted value of y_train
```

```python
plt.scatter(x_train, y_train, color='lightcoral')
plt.plot(x_train, y_pred_train, color='blue')
plt.title("Experience vs Salary (Training Set)")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['x_train/Pred(y_test)', 'x_train/y_train'], title='Sal/Exp', loc='best',facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(x_test,y_test, color='lightcoral')
plt.plot(x_train, y_pred_train, color='blue')
plt.title("Years of Experience vs Salary (Test Set)")
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['x_train/Pred(y_test)','x_train/y_train'], title='Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
print(f"Coefficient: {reg.coef_}")
print(f"Intercept: {reg.intercept_}")
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred_test)
print(f"Mean Absoulute Error: {mae}")
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred_test)
print(f"Mean Squared Error: {mse}")
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_test,y_pred_test)
print(f"Root Mean Squared Error: {rmse}")
```
